```python
import csv
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

  
```

```python
df = pd.read_table("ip_latencies.csv", sep=",")
```

Clear Nan latencies


```python
df = df[df['latency'].notnull()]
```

```python
df
```

```python
df = df.groupby('ip')['latency'].apply(list)
```

```python
def density_plot(all_diff: dict):
    plt.figure(figsize=(15, 15))

    for ip, data in all_diff.items():
        sns.distplot(data, hist=False, kde=True,
                     kde_kws={'linewidth': 3},
                     label=ip)

    # Plot formatting
    plt.legend(prop={'size': 16}, title='IP')
    plt.title('Density Plot of network latencies')
    plt.xlabel('Delay (ms)')
    plt.ylabel('Density')
    plt.show()
```

```python

density_plot(df.to_dict())
```
